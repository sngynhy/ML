# 범주형 컬럼 LabelEncoding처리

```python
cate_cols=['workclass','education','marital-status','occupation','relationship','race','gender','navtive-country', 'income']
not_cate_cols=['age', 'fnlwgt','education-num','capital-gain', 'capital-loss', 'hours-per-week']
```

```python
adult.columns
```

```python
adult
```

```python
le.classes_
```

```python
# series를 받아서 LabelEncoding 처리하는 함수
from sklearn.preprocessing import LabelEncoder

col_dic={}
def encoding(x):
    le=LabelEncoder()
    label_x=le.fit_transform(x)  # 학습 & 변환 동시에 처리
    col_dic[x.name]=le.classes_
    return label_x
    
```

```python
df=adult[cate_cols].apply(encoding)
df
```

```python
col_dic['occupation']
```

```python
adult_df=adult[not_cate_cols].join(df)
adult_df
```

```python
# 데이터 셋 분리(X, y 분리 -> train, test 분리)
y=adult_df['income']
X=adult_df.drop(columns='income')
```

```python
# train, test 분리
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test=train_test_split(X, y)
```

```python
# 모델 생성 및 학습
tree=DecisionTreeClassifier(random_state=1)
tree.fit(X_train, y_train)
```

```python
# 예측 및 평가
pred_train=tree.predict(X_train)
pred_test=tree.predict(X_test)
```

```python
accuracy_score(y_train, pred_train)
```

```python
accuracy_score(y_test, pred_test)
```
